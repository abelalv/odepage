```python
import numpy as np
from scipy import integrate


from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation

from matplotlib import pyplot as plt

# Funcion de flujo
def Flow(x_y, t0, a,b,c,d):
    x, y = x_y
    return [a*x+b*y, c*x+d*y]
# plot of eigenvectors  
def eig_vectors(a):    
    x, v = np.linalg.eig(a)
    if np.trace(a)**2-4*np.linalg.det(a)<0:
        plt.text(0.5*v[0,0], 0.5*v[1,0], '$\lambda=$'+ str(round(x[0],1)), fontsize=20, color='red')
    else:
        zz=max(abs(x))
        v[:,0]
        #plt.quiver(v[0,0],v[1,0],scale=1, pivot='mid',color='red')
        plt.plot([0,x[0]*v[0,0]/zz],[0,x[0]*v[1,0]/zz],color='red')
        plt.plot([0,x[1]*v[0,1]/zz],[0,x[1]*v[1,1]/zz],color='green')
        #plt.quiver(v[0,1],v[1,1],scale=1, pivot='mid',color='green')
        plt.text(0.5*v[0,0], 0.5*v[1,0], '$\lambda=$'+ str(round(x[0],1)), fontsize=20, color='red')
        plt.text(0.5*v[0,1],0.5*v[1,1], '$\lambda=$'+ str(round(x[1],1)), fontsize=20, color='green')
    return

    #plt.text(0.55, 0.55, 'h', fontsize=20, color='green')
def ceroclinas(a):
    x=np.linspace(-2,2,100)
    y1=(-a[0,0]*x)/a[0,1]
    y2=(-a[1,0]*x)/a[1,1]
    plt.plot(x,y1,color='black')
    plt.plot(x,y2,color='black')
    return
    

def campo(x_max,x_min,y_max,y_min,a,b,c,d):

    # Grid de puntos (x, y) en el plano
    x = np.linspace(x_min, x_max, 25)
    y = np.linspace(y_min, y_max, 25)
    xx, yy = np.meshgrid(x, y)
    uu, vv = Flow((xx, yy),0,a,b,c,d)
    norm = np.sqrt(uu**2 + vv**2)
    uu = uu / norm
    vv = vv / norm
    #plt.quiver(xx, yy, uu, vv, norm, cmap=plt.cm.gray)
    #plt.xlabel('Compradores')
    #plt.ylabel('Vendedores')
    #plt.show()
    return xx, yy, uu, vv, norm
```

```python
import warnings
warnings.filterwarnings("ignore")
def df_dt(min_time,max_time,a0,b0,c0,d0):
    N=16  # number of trajectories to plot
    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1])
    ax.axis('on')


    # prepare the axes limits
    ax.set_xlim((-2, 2))
    ax.set_ylim((-2, 2))
    

    # Choose random starting points, uniformly distributed from -15 to 15
    np.random.seed(1)
    #x0 = np.zeros(N,2) + 3* np.random.random((N, 2))
    x0=np.zeros([N,2])
    for n in range(N):
        x0[n,:] = np.array([0.5*np.cos(n*np.pi/(8)),0.5*np.sin(n*np.pi/(8))])
        
    #x0=np.array([1.0,1.0])
    # Solve for the trajectories
    t = np.linspace(min_time, max_time, 100)
    y = np.asarray([integrate.odeint(Flow, x0, t,args=(a0,b0,c0,d0))
                      for x0 in x0])
   
    #y=integrate.odeint(sistema_ode, x0, t)
    # choose a different color for each trajectory
    #colors = plt.cm.jet(np.linspace(0, 1, N))
    for i in range(N):
        lines = ax.plot(y[i,:,0], y[i,:,1], '-', c='blue')
        plt.setp(lines, linewidth=2)
    xx, yy, uu, vv, norm=campo(-2,2,-2,2,a0,b0,c0,d0)
    plt.quiver(xx, yy, uu, vv, norm, cmap=plt.cm.gray)
    #plt.plot(y[1,:,],y[1,:,1])
    # calculation of the jacobian matrix
    Df=np.array([[a0,b0],[c0,d0]])
    # plot of the eigenvectors
    
    eig_vectors(Df)
    #ceroclinas(Df)
    plt.grid(True)
    plt.show()
    #warnings.simplefilter("ignore")
    return y

'''



    for i in range(N):
        x, y, z = x_t[i,:,:].T
        lines = ax.plot(x, y, z, '-', c=colors[i])
        plt.setp(lines, linewidth=2)
''' 
    #
    #ax.view_init(30, angle)
    #plt.show()
```
